# Lecture 3 Part 2

```r
# Libraries
library(readr)
library(dplyr)
```

```r
# Import data
heart_df <- readr::read_csv(datafile)
```

## Section 1

```r
# Determine the unique elements for the smoking variable.
heart_df %>%
  distinct(smoking)
```

## Section 2

```r
# Select all the observations for the smoking variable only.
heart_df %>%
  select(smoking)
```

## Section 3

```r
# Select all the observations only for those who do not smoke.
heart_df %>%
  filter(smoking == "No")
```

## Section 4

```r
# Select the observations in rows 100 through 110 and show only the diabetes and smoking variable data.
heart_df %>%
  slice(100:110) %>%
  select(diabetes, smoking)
```

## Section 5

```r
# Select all the observations for those who smoked and died.
heart_df %>%
  filter(smoking == "Yes" & death == "Yes")
```

## Section 6

```r
# Select all the observations for those who were either female or who did not smoke or who has diabetes.
heart_df %>%
  filter(sex == "Female" | smoking == "No" | diabetes == "Yes")
```
