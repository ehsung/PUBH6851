# Problem 4.1

```python
import sys
import math
import csv
import pandas as pd

# Uncomment this code to use terminal calling with sys.argv
# numbers = [float(arg) for arg in sys.argv[1:]]

# This uses in code user input instead
user_input = input("Enter numbers separated by a space: ")

# Split the string into a list of individual numbers
numbers = user_input.split()

# Convert the list of strings to a list of numbers
numbers = [float(number) for number in numbers]
print("You entered: ", numbers)

# Open the CSV file for writing
with open('results.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile, delimiter=';')

  # Write the header row
  writer.writerow(['Number', 'Log', 'Square', 'Sine', 'Cosine'])

  # Used to generate a dataframe version (if you don't want to open results.csv)
  df_results = []

  # Calculate and write the results for each number
  for number in numbers:
    log_value = math.log(number)
    square_value = number ** 2
    sine_value = math.sin(number)
    cosine_value = math.cos(number)
    writer.writerow([number, log_value, square_value, sine_value, cosine_value])
    df_results.append([number, log_value, square_value, sine_value, cosine_value])

# Create a Pandas DataFrame from the results
df = pd.DataFrame(df_results, columns=['Number', 'Log', 'Square', 'Sine', 'Cosine'])
df
```

```python
import numpy as np

# Step 1
# Create a 10x10 matrix with 1-100
mat = np.arange(1, 101).reshape(10, 10)


# Step 2
def num_elements(m):
  '''
  Return number of elemnts in a given matrix (numpy array) m
  '''
  return m.size

def num_rows(m):
  '''
  Return number of rows in a given matrix (numpy array) m
  '''
  return m.shape[0]

def num_columns(m):
  '''
  Return number of columns in a given matrix (numpy array) m
  '''
  return m.shape[1]


# Step 3
def row_means(m):
  '''
  Return array of means of the rows for a matrix m
  '''
  return np.mean(m, axis=1)

def column_means(m):
  '''
  Return array of means of the columns for a matrix m
  '''
  return np.mean(m, axis=0)

# Compute means for mat
row_means_mat = row_means(mat)
column_means_mat = column_means(mat)

print("row_means_mat:", row_means_mat)
print("")
print("column_means_mat:", column_means_mat)
print("")


# Step 4
# Create array b, exactly 50 points equally spaced in the range from 11 to 23
b = np.linspace(11, 23, 50)

# Calculate sum of squares using list comphrension
print("sum of squares in array b: ", sum([x**2 for x in b]))
```

# Problem 4.3

```python
import numpy as np
from collections import OrderedDict

# Step 1
# Toy Data from Homework
geneList = np.array(["A2M", "FOS", "BRCA2", "CPOX"])
values0 = np.array([0.12, 0.08, 0.06, 0.02])
values1 = np.array([0.01, 0.07, 0.11, 0.09])
values2 = np.array([0.03, 0.04, 0.04, 0.02])
values3 = np.array([0.05, 0.09, 0.11, 0.14])

toy_data_combined = np.array([values0, values1, values2, values3])
print("4x4 Numpy Array:")
print(toy_data_combined)
print()


# Step 2
mean_expressions = np.mean(toy_data_combined, axis=1)

# Create the dictionary
gene_mean_expression = dict(zip(geneList, mean_expressions))
print("Mean Gene Expression Dictionary:")
print(gene_mean_expression)
print()


# Step 3
# Find the key (gene) with largest mean expression
largest_gene = max(gene_mean_expression, key=gene_mean_expression.get)

# Find the value (expression) that is the largest
largest_mean_expression = gene_mean_expression[largest_gene]

print(largest_gene, "has the largest mean expression of", largest_mean_expression)
print()


# Step 4
# Sort using sorted in decending order (reverse) of the expression level for the dict
sorted_gene_mean_expression = dict(sorted(gene_mean_expression.items(), key=lambda item: item[1], reverse=True))
print("Sorted Gene Mean Expression:")
print(sorted_gene_mean_expression)
```

# Problem 4.4

```python
# process_cc.py
import csv,os,sys,pickle
import numpy as np

txtFile = "/content/cellcycle.txt"        # make sure that you have correct path to cellcycle.txt here

## read the file once to get numRows and numCols
numRows = 0
reader = csv.reader(open(txtFile, 'r'), delimiter='\t')
expListIDs = next(reader)
expListIDs = np.array(expListIDs[1:])
for linja in reader:
    numRows += 1

## populate a matrix and name vectors with file info
numColumns = len(expListIDs)
exprMat = np.zeros([numRows,numColumns])
reader = csv.reader(open(txtFile, 'r'), delimiter='\t')
header = next(reader)
rowInd = 0
geneList = []

for linja in reader:
    row = np.array(linja[1:])
    newRow = np.zeros(len(row),)
    nanInds = np.where(row == '')
    goodInds = np.where(row != '')
    newRow[nanInds] = np.nan
    newRow[goodInds] = [float(element) for element in row[goodInds]]
    exprMat[rowInd,:] = newRow
    rowInd +=1
    geneList.append(linja[0])
geneList = np.array(geneList)

## print out info
print("--------------")
print(f"matrix of size {exprMat.shape} created")
print(f"gene list size:\t{geneList.size}")
print(f"exp list size:\t{expListIDs.size}")
print("--------------")

## write the data to a file
outFilePath = os.path.join(".","excercise.pickle")
tmp = open(outFilePath,'wb')
pickle.dump([geneList,expListIDs,exprMat],tmp)
tmp.close()
```

```python
import pickle
import matplotlib.pyplot as plt

# Step 1
# Open created pickle file
with open('/content/excercise.pickle', 'rb') as file:
    pickled_data = pickle.load(file)

# Access the arrays
geneList = pickled_data[0]
expListIDs = pickled_data[1]
exprMat = pickled_data[2]


# Step 2
# Initialize condition_max_geneList
condition_max_geneList = []

# Loop through the columns in exprMat, which corresponds to the conditions
for i in range(exprMat.shape[1]):
  # Find the max value in column and its respective row, which corresponds to the gene
  max_row_index = np.argmax(exprMat[:, i])

  # Append the gene with the maximum expression value for the looped condition (column)
  condition_max_geneList.append(geneList[max_row_index])

# There are 82 values in this list, each correspeonding to a condition and its max expression gene name
print(condition_max_geneList)
print()


# Step 3
def gene_expression_hist(gene):
  '''
  Plots a histogram based on gene name and its respective expression values across conditions
  '''
  # Create a gene index dictionary that can be used for the exprMat
  gene_index_dict = {gene: index for index, gene in enumerate(geneList)}

  if gene in gene_index_dict:
    # Get index
    gene_index = gene_index_dict[gene]

    # Plot histogram
    plt.hist(exprMat[gene_index], bins='auto')

    # Add labels and title
    plt.xlabel('Expression Values')
    plt.ylabel('Counts')
    plt.title(f"Histogram of Expression Values for {gene}")
  else:
    # If the gene is not in geneList, then print not found
    print(f"Gene '{gene}' not found in geneList.")


gene_expression_hist("aa")
print()
gene_expression_hist("YBL045C")
```

```python
# Step 4
# Same code as Step 3, but changed plt.hist to plt.plot for line graph
def gene_expression_line(gene):
  '''
  Plots a line graph based on gene name and its respective expression values across conditions
  '''
  # Create a gene index dictionary that can be used for the exprMat
  gene_index_dict = {gene: index for index, gene in enumerate(geneList)}

  if gene in gene_index_dict:
    # Get index
    gene_index = gene_index_dict[gene]

    # Plot histogram
    plt.plot(expListIDs, exprMat[gene_index], marker='o')

    # Add labels and title
    plt.xlabel('Condition')
    plt.xticks(rotation='vertical')
    plt.xticks(fontsize=5)
    plt.ylabel('Expression Value')
    plt.title(f"Line Graph of Expression Values for {gene}")
  else:
    # If the gene is not in geneList, then print not found
    print(f"Gene '{gene}' not found in geneList.")

gene_expression_line("aa")
print()
gene_expression_line("YBL045C")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data
data = np.loadtxt('/content/RADIOLOGY.txt')

# Define column labels
column_labels = ['visits to radiology', 'patient-days,', 'ER visits', 'clinic visits']

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Loop through columns in data and plot
for i, ax in enumerate(axes.flatten()):
  column_data = data[:, i + 1] # Skip first column, which is just the index
  mean_value = np.mean(column_data)

  # Plot in subplot
  ax.plot(column_data, 'bo')
  ax.axhline(y=mean_value, color='r', linestyle='-')

  # Add labels and title to the subplot to match example
  ax.set_title(f'{column_labels[i]}')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
