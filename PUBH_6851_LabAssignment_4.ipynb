```r
# Install Packages
install.packages("ggthemes")
install.packages("ggmosaic")
```

```r
# Libraries
library(readr)
library(dplyr)
library(ggplot2)
library(ggthemes)
```

```r
# Import Data
plottingdata = read_csv("/content/PlottingData.csv")
```

## Section 1
Select one numerical and one categorical variable in the data set imported in this notebook and generate a box-and-whisker plot that has not been created as example in this notebook.

```r
plottingdata %>% ggplot(
  aes(
    x = Ventilation,
    fill = Ventilation,
    y = dBP
  )
) +
  geom_boxplot() +
  labs(
  title = "dBP distribution of participants",
  subtitle = "Grouped by Ventilation"
  ) +
  ggthemes::theme_economist()
```

## Section 2
Select two numerical variables and a categorical variable and create a scatter plot (grouped by the levels of the categorical variable) that has not been created as example in this notebook.

```r
plottingdata %>% ggplot(
  aes(
    x = dBP,
    y = HB
    )
  ) +
    geom_point(
    aes(
      color = Ventilation,
      shape = Ventilation
    ),
    size = 3
  ) +
  labs(
    title = "dBP vs HB",
    subtitle = "Grouped by Ventilation"
  ) +
  ggthemes::theme_economist()
```

## Section 3
Look up what a mosaic plot is. Import an appropriate library and create a mosaic plot of the variables
`Ventilation` and `Obesity`.

```r
library(ggmosaic)

plottingdata %>%
  ggplot() +
  geom_mosaic(
  aes(
    x = product(Ventilation),
    fill = Obesity)
    ) +
  labs(
    title = "Mosaic Plot of Ventilation and Obesity"
  ) +
  ggthemes::theme_economist()
```
